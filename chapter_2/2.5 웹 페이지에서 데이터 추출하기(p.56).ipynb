```python
# 2.5.1 정규 표현식으로 스크레이핑하기(re모듈 사용)
import re
```

```python
# re.search() 함수를 사용하면 두 번째 매개변수의 무나열이 첫 번째 매개변수의 정규 표현식에 맞는지 확인할 수 있음
# 맞는 경우 Match 객체를 반환 않는다면 None을 반환
# match = 'abc'를 보면 abc가 매치된 것을 확인할 수 있습니다.
re.search(r'a.*c', 'abc123DEF')
```

```python
re.search(r'a.*d', 'abc123DEF')
```

```python
# re.IGNORECASE(또는 re.I)를 지정하면 대소문자를 무시함
re.search(r'a.*d','abc123DEF', re.IGNORECASE)
```

```python
# Match 객체의 group() 메서드로 일치한 값을 추출함
# 매개변수에 0을 지정하면 맻된 모든 값을 반환함
m = re.search(r'a(.*)c', 'abc123DEF')
m.group(0)
```

```python
# 매개변수에 1 이상의 숫자를 지정하면 정규 표현식에서 ()로 감싼 부분에 해당하는 값을 추출
m.group(1)
```

```python
# re.findall() 함수를 사용하면 정규 표현식에 맞는 모든 부분을 추출할 수 있음
# \w는 유니코드로 글자를 비교. 이 밖에도 공백 문자는 \s 등으로 추출할 수 있음
# 다음 예는 2글자 이상의단어를 모두 추출
re.findall(r'\w{2,}', 'This is a pen')
```

```python
# re.sub() 함수를 사용하면 정규 표현식에 맞는 부분을 바꿀 수 있음
# 3번째 매개변수에 넣은 문자열에서 첫 번째 정규 표현식에 맞는 부분을 2번째 매개변수로 변경
re.sub(r'\w{2,}', 'That', 'This is a pen')
```
